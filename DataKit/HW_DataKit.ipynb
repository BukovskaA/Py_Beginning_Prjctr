- Tabulate the following function: F(x)=2x^2+5, x∈[1,100] with step 1.

```python
from tabulate import tabulate

def F(x):
    return 2 * x ** 2 + 5

start_x = 1
end_x = 100
step = 1

x_values = np.arange(start_x, end_x + 1, step)

F_x_values = F(x_values)

table_data = np.column_stack((x_values, F_x_values))

headers = ["x", "F(x)"]

table = tabulate(table_data, headers, tablefmt="fancy_grid", numalign="center", stralign="center")

print(table)
```

- Tabulate the following function: F(x)=e^−x, x∈[−10,10] with step 1.

```python
from tabulate import tabulate

start_x = -10
end_x = 10
step = 1

x_values = np.arange(start_x, end_x + 1, step)

F_x_values = np.exp(-x_values) # Calculating F(x) using np.exp(a math function used to calculate the exponential values(a number is equivalent to the number being multiplied by itself a particular set of times) of all the elements present in the input array. )

table_data = np.column_stack((x_values, F_x_values))

headers = ["x", "F(x)"]

table = tabulate(table_data, headers, tablefmt="fancy_grid", numalign="center", stralign="center", floatfmt=(".0f", ".4f"))

print(table)
```

DataViz

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

car_crashes = sns.load_dataset("car_crashes")

car_crashes.head()
```

Distribution

```python

fig, axs = plt.subplots(ncols=3)
fig.set_size_inches(16, 5)

# sns.displot(car_crashes, x='speeding', kind = 'kde')

# Alternative for sns.displot, which cannot be used with custom axes, since it's figure-level
sns.histplot(car_crashes['speeding'], bins=10, ax=axs[0])
sns.kdeplot(car_crashes['speeding'], fill=True, bw_adjust=.5, cut=0, ax=axs[1])

# Empirical cumulative distributions
_ = sns.ecdfplot(car_crashes['speeding'], ax=axs[2])
```

Regression plot


```python
_ = sns.regplot(x ='total', y ='alcohol',  data = car_crashes)
```

Categorical plots

```python
# Creating a categorical plot using catplot


plt.figure(figsize=(10, 8))
_ = sns.catplot(y="abbrev", x="total", data=car_crashes, height=6, kind="bar", palette="viridis")
plt.title("Total Crashes by States")
plt.xlabel("Total Crashes")
plt.ylabel("State Abbreviation")
plt.show()
```

Scatter plot

```python
penguins = sns.load_dataset("penguins")

penguins.head()
```

```python
# stripplot 
_ = sns.catplot(y="species", x="flipper_length_mm", data=penguins, kind='strip', palette='Set2')
```

```python
sns.relplot(
    data=penguins,
    x="bill_length_mm", y="bill_depth_mm", hue="body_mass_g"
)
```

```python
# swarm plot to visualize the distribution of flipper lengths by species
plt.figure(figsize=(10, 6))
sns.swarmplot(x="species", y="flipper_length_mm", data=penguins, palette="Set2")
plt.title("Swarm Plot of Flipper Length by Penguin Species")
plt.xlabel("Species")
plt.ylabel("Flipper Length (mm)")
plt.show()
```
