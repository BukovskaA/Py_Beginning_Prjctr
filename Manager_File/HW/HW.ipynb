```python
"""
1. This code generates 26 text files named A.txt, B.txt, and so on up to Z.txt,
appends a random number between 1 and 100 to each file, and creates a summary file
(summary.txt) that contains the name of the file and the number in that file.
"""

import random

def generate_files():
    for letter in range(ord('A'), ord('Z')+ 1):
        file_name = chr(letter) + ".txt"
        # creating a file object for the file
        with open(file_name, "w") as f:
            # random numbers to the file
            f.write(str(random.randint(1, 100)))

# generationg a summarry file

def generate_summary():
    summary_file = open("summary.txt", "w")
    #iterationg
    for letter in range (ord('A'), ord('Z')+ 1):
        file_name = chr(letter) + ".txt"
        number = random.randint(1, 100)
        #writing the file name to the summary file
        summary_file.write(f"{file_name}:{number}/n")
    summary_file.close()

if __name__ == "__main__":
    generate_files()
    generate_summary()

# 2nd variant

import random
import string # module to use ascii_uppecase

def generate_random_number():
    return random.randint(1, 100)

def generate_text_files():
    letters = list(string.ascii_uppercase)
    with open("summary.txt", "w") as summary_file:
        for letter in letters:
            filename = f"{letter}.txt"
            with open(filename, "a") as file:
                number = generate_random_number()
                file.write(str(number))
            summary_file.write(f"{filename}: {number}\n")

if __name__ == "__main__":
    generate_text_files()

with open("summary.txt", "r") as summary_file:
    print(summary_file.read())
```

```python
# 3. Write a program that will simulate user scores in a game.


import random
import csv


def simulate_scores():
    # Creating a list of players
    players = ["Josh", "Luke", "Kate", "Mark", "Mary"]

    # Creating a list to store the scores
    scores = []

    # Iterating through the players list
    for player in players:
        # Initializing the score to 0
        score = 0

        # Iterationg through 100 rounds
        for round in range(100):
            # Adding a random number between 0 and 1000 to the score
            score += random.randint(0, 1000)

        # Adding the player's name and score to the scores list
        scores.append((player, score))

    # Return the scores list
    return scores


def save_scores(scores):
    # Opening the scores.csv file in write mode
    with open("scores.csv", "w") as csvfile:

        # Creating a csv writer object
        writer = csv.writer(csvfile, delimiter=",")

        # Writing the header row
        writer.writerow(["Player name", "Score"])

        # Writing the scores to the file
        for player, score in scores:
            writer.writerow([player, score])


scores = simulate_scores()

save_scores(scores)
```

```python
import csv
from operator import itemgetter

def read_scores_from_csv(filename):
    # Read scores from the specified CSV file and store them in a dictionary
    with open(filename, "r") as csvfile:
        reader = csv.reader(csvfile)
        next(reader)  # Skip the header row
        scores = {row[0]: int(row[1]) for row in reader}
    return scores

def save_high_scores(scores):
    # Save the high scores to a new CSV file in descending order
    with open("high_scores.csv", "w", newline="") as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(["Player name", "Highest score"])
        # Sort the scores dictionary by value (the score) in descending order
        for player, score in sorted(scores.items(), key=itemgetter(1), reverse=True):
            writer.writerow([player, score])

scores = read_scores_from_csv("game_scores.csv")
save_high_scores(scores)
```
